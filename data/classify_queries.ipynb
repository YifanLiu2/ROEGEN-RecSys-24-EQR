```python
# Define a function to classify and group queries
def classify_queries(file_path):
    # Initialize a dictionary to hold categories and their corresponding queries
    categories = {}
    
    # Open the file for reading
    with open(file_path, 'r', encoding='utf-8') as file:
        # Iterate over each line in the file
        for line in file:
            # Strip whitespace and split the line into query and category
            parts = line.strip().split(" - ")
            if len(parts) == 2:
                query, category = parts
                # Add the query to the appropriate category list in the dictionary
                if category not in categories:
                    categories[category] = [query]
                else:
                    categories[category].append(query)
    
    # Return the dictionary of categories and queries
    return categories

# Path to the file containing the queries
file_path = "final_queris_v0.txt"

# Call the function and store the result
classified_queries = classify_queries(file_path)

# Print the classified queries
for category, queries in classified_queries.items():
    print(f"{category}:")
    for query in queries:
        print(f"- {query}")
    print()  # Print a newline for better readability between categories
```

```python
import json
file_path = 'output/new_test/processed_query_q2d.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Variables for accuracy calculation
correct_classifications = 0
total_classifications = 0

# Lists to store wrong classifications
false_positives = []
false_negatives = []

# Check the classification of each query
for entry in data:
    query = entry['Query']
    classified_broad = len(entry['broad']) > 0
    expected_broad = query in Broad_and_Indirect
    
    if classified_broad == expected_broad:
        correct_classifications += 1
    else:
        if classified_broad:
            false_positives.append(query)
        else:
            false_negatives.append(query)
    
    total_classifications += 1

# Calculate accuracy
accuracy = correct_classifications / total_classifications

print(f"Accuracy: {accuracy:.4f}")

# Print wrong classifications
print("False Positives (classified as broad but are not):")
for query in false_positives:
    print(f"  {query}")

print("False Negatives (not classified as broad but should be):")
for query in false_negatives:
    print(f"  {query}")

# Function to automate the checking process
def check_classification(data):
    results = []
    for entry in data:
        query = entry['Query']
        classified_broad = len(entry['broad']) > 0
        expected_broad = query in Broad_and_Indirect
        is_correct = classified_broad == expected_broad
        
        result = {
            'Query': query,
            'Classified as Broad': classified_broad,
            'Expected Broad': expected_broad,
            'Correct Classification': is_correct
        }
        results.append(result)
    return results

# Print detailed results
results = check_classification(data)
for result in results:
    print(result)
```
